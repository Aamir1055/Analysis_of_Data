```python
# Import the necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
gas = pd.read_csv("gas_prices.csv")
plt.figure(figsize=(10,7))
plt.title("Gas price over time (in USD)")
plt.plot(gas.Year,gas.USA,'b.-',label ='United States')
plt.plot(gas.Year,gas["Canada"],'r.-',label ='Canada')
plt.plot(gas.Year,gas["South Korea"],'g.-',label='South Korea')
print(gas.Year[::3])
plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.legend()
plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

```python
bins = [40,50,60,70,80,90,100]

plt.figure(figsize=(8,5))

plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

#plt.savefig('histogram.png', dpi=300)

plt.show()
```

```python
import matplotlib.pyplot as plt
  
# line 1 points
x1 = [1,2,3]
y1 = [2,4,1]
# plotting the line 1 points 
plt.plot(x1, y1, label = "line 1")
  
# line 2 points
x2 = [1,2,3]
y2 = [4,1,3]
# plotting the line 2 points 
plt.plot(x2, y2, label = "line 2")
plt.title('Two lines on same graph!')
   
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
# giving a title to my graph
plt.title('Two lines on same graph!')
  
# show a legend on the plot
plt.legend()
  
# function to show the plot
plt.show()
```

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
right
```

```python
labels = ('Left','Right')
colors = ('#abcdef','#aabbcc')
plt.pie([left,right],labels=labels,colors=colors, autopct='%.2f %%')
plt.title("Foot prefreence of FIFA players")
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
plt.style.use('ggplot')
light = fifa.loc[fifa.Weight<125].count()[0]
light_medium =fifa.loc[(fifa.Weight>=125) & (fifa.Weight<150)].count()[0]
medium = fifa.loc[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight>=175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight>=200].count()[0]
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['under 125','125-150','150-175','175-200','over 200']
explode = [.4, .1, .1, 0, 0.4]
plt.pie(weights,labels=labels,explode=explode,autopct='%.2f %%',pctdistance=0.8)
plt.title("Weights distribution of fifa players (in lbs)")
plt.show()
```

```python
plt.figure(figsize=(5,8))
plt.style.use('default')
barcelona = fifa.loc[fifa.Club=='FC Barcelona']['Overall']
rev=fifa.loc[fifa.Club=='New England Revolution']['Overall']
madrid = fifa.loc[fifa.Club=='Real Madrid']['Overall']
#labels=['FC Barcelona','Real Madrid','New England Revolution']
plt.ylabel('FIFA overall rating')
plt.title("Professional Football team comparision")
bp=plt.boxplot([barcelona,madrid,rev],labels=['FC Barcelona','Real Madrid','New England Revolution'], patch_artist=True)
for box in bp['boxes']:
      box.set(color='blue', linewidth=2)
      box.set(facecolor =  '#e0e0e0' )

plt.show()
```
